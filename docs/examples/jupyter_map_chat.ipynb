# Jupyter Map Chat GUI

This notebook shows the ipywidgets-based GeoAgent chat panel for an existing `leafmap.Map` or `anymap.Map` object.

- The map renders on the left.
- The chat panel renders on the right.
- Use **Hide chat** to collapse the right panel and let the map take the full notebook width.
- Set `OPENAI_API_KEY` or choose another configured provider before sending prompts.

````python
# %pip install -q "GeoAgent[jupyter,leafmap,openai]"
````

## leafmap

````python
import leafmap.maplibregl as leafmap
from geoagent.ui import map_chat

m = leafmap.Map(center=[-100, 40], zoom=4)
map_chat(m, provider="openai-codex", model_id="gpt-5.5")
````

Try prompts such as:

- `Add a marker for Knoxville and zoom to it.`
- `List the current map layers.`
- `Change the basemap to OpenTopoMap.`
